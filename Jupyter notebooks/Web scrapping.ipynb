```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage=requests.get('https://www.flipkart.com/search?q=camera&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off').text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
final=pd.DataFrame()
for i in range(1,5):
    url='https://www.flipkart.com/search?q=camera&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}'.format(i)
    webpage=requests.get(url).text
    soup=BeautifulSoup(webpage,'lxml')
    titles=[]
    prices=[]
    images=[]
    for d in soup.find_all('div',attrs={'class':'_2kHMtA'}):
        title=d.find('div',attrs={'class':'_4rR01T'})
        price=d.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
        image=d.find('img',attrs={'class':'_396cs4'})
        titles.append(title)
        prices.append(price)
        images.append(image)
    d={'title':titles,'price':prices,'image':images}
    df=pd.DataFrame(d)
    final = pd.concat([final, df], ignore_index=True)
    
```
